Теперь сравним то, как часто летают обладатели бонусной карты и те, у кого ее нет.

```python
with_bonus = df.loc[df['ProgrammNumber'].notnull()]
x1 = with_bonus.groupby(by=['PassengerDocument'], dropna=True)['NumberFlight'].count()
```

```python
right = df[['Last__Name', 'First__Name', 'PassengerMiddleName', 'PassengerDocument']]
right
```

```python
no_bonus = df.loc[df['ProgrammNumber'].isnull()]
x2 = no_bonus.groupby(by=['PassengerDocument'], dropna=True)['NumberFlight'].count()
x2.value_counts()
```

```python
x2.shape
```

```python
m = pd.merge(x2, right, on = 'PassengerDocument')
m
```

```python
import pandas as pd
import numpy as np
```

### Load dataset

```python
df = pd.read_csv('data/BoardingData.csv', sep=';')
```

```python
df.head()
```

### Создаем поле для каждого полета

```python
df['FlightNumber_by_date'] = df['FlightNumber'] + ' ' + df['FlightDate'] + ' ' + df['FlightTime']
```

### Фильтруем только мужчин

```python
df = df[df['PassengerSex'] == 'Male']
df.head()
```

### Создаем список полетов для каждого пассажира

```python
passenger_flights = {}

for index, row in df[['PassengerDocument', 'FlightNumber_by_date']].iterrows():
    document = row['PassengerDocument']
    flight_number = row['FlightNumber_by_date']
    
    if not document in passenger_flights:
        passenger_flights[document] = []
    passenger_flights[document].append(flight_number)
        
passenger_flights
```

### Вычисляем пассажиров у которых больше одного полета

```python
passenger_flights_more_one = {}

for key, value in passenger_flights.items():
    if len(value) > 1:
        passenger_flights_more_one[key] = value
        
passenger_flights_more_one
```

### Находим пассажиров, которые летали вместе

```python
result_list = []

for doc, number in passenger_flights_more_one.items():
    for doc1, number1 in passenger_flights_more_one.items():
        if doc != doc1:
            intersect = set(number).intersection(set(number1))
            if len(intersect) > 1:
                first = str(df[df['PassengerDocument'] == doc]['PassengerLastName'].unique())
                second = str(df[df['PassengerDocument'] == doc1]['PassengerLastName'].unique())
                if first.strip()[:7] != second.strip()[:7]:
                    result_list.append([first, doc, second, doc1, intersect])
                    print(first, second)
                    print(doc, ':', doc1, ':', intersect)
                    print('-' * 20)
                    
```
